# Collecting Samples
My new idea is to collect samples of different types of roofs, grass, trees, water, dirt, etc. and compute the color histograms for these different classes. Then maybe we can use template matching to find candidate roof locations, then take local color histograms and classify those image patches.

Let's start with roofs and water and see how we can visualize the difference between their color histograms.

```python
from color_histograms import *
import IPython.display
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
imageDir = '../images/'
samplesDir = imageDir + 'samples/'
roofsDir = samplesDir + 'roofs/'
waterDir = samplesDir + 'water/'
vegDir = samplesDir + 'vegetation/'
```

I'm going to use gThumb to crop and save samples. Sample images will be saved with maximum (9) compression to save space since it doesn't affect the pixel values.
